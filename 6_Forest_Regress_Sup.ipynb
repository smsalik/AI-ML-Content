## Forest Regression

```python
#   Random Forest algorithm is a powerful tree learning technique in Machine Learning 
#   to make predictions and then we do voting of all the tress to make prediction. 
#   They are widely used for classification and regression task.

#   It is a type of classifier that uses many decision trees to make predictions.
#   It takes different random parts of the dataset to train each tree and then 
#    it combines the results by averaging them. This approach helps improve the accuracy of predictions. 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\Admin\Datasets\Position_Salaries.csv")
```

```python
dataset
```

```python
x = dataset.iloc[: , 1:-1].values
y = dataset.iloc[: , -1].values
```

```python
x , y
```

```python
# Now train our model
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
```

```python
regressor.fit(x,y)
```

```python
# By above my model is trained now
```

```python
# Now we will do prediction
```

```python
regressor.predict([[6.5]])
```

```python
# Random Forest and Decision Tress are good models and algorithm for the
# greater number of feautures. They do not give good prediction for lower number of features.
```

```python
# To Visualize it, we will increase the number of grids
```

```python
x_grid = np.arange(min(x), max(x), 0.01)
```

```python
x_grid.shape
```

```python
x_grid = x_grid.reshape(len(x_grid),1)
```

```python
x_grid.shape
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid), color='blue')
plt.title('Random Tree Regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
