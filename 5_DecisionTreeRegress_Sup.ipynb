## Decision Tree Regression

```python
#  Decision Tree Regression is a non-linear regression method that can handle complex 
#  datasets with intricate patterns. It uses a tree-like model to make predictions, 
#  making it both flexible and easy to interpret.
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\Admin\Datasets\Position_Salaries.csv")
```

```python
dataset
```

```python
x = dataset.iloc[: , 1:-1].values
y = dataset.iloc[: , -1].values
```

```python
x , y
```

```python
# Now we import our model here
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
```

```python
# Now fit our model
```

```python
regressor.fit(x , y)
```

```python
# Decision tree regressor is for complex Datatype 
```

```python
# Now predicting through our model
```

```python
regressor.predict([[6.5]])
```

```python
# Results are not good, as Decision Tree is for Complex dataset.
# you can see 6.5 give 150000, which is already value of 6 = 150000
# It must give higher value
```

```python
x_grid = np.arange(min(x), max(x), 0.01)
```

```python
x_grid.shape
```

```python
x_grid = x_grid.reshape(len(x_grid),1)
```

```python
x_grid.shape
```

```python
# Now it is in two dimension array
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid), color = 'blue')
plt.title('Decision Tree Regressor')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
