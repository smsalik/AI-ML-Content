### Split the dataset into 3 subset Numerical(num) , Categorical(cat) and Datetime

```python
num = df.select_dtypes([int, float])

cat = df.select_dtypes([object])

datee = df.select_dtypes([np.datetime64])
```

### Transform LoanStatus column to binary values : 1 = Defaulted and 0 (all other values)

```python
cat['LoanStatus'].unique()
```

```python
def transform_to_binary(x):
    if x == 'Defaulted':
        return 1
    else:
        return 0

cat['LoanStatus'] = cat['LoanStatus'].apply(transform_to_binary)
```

```python
cat['LoanStatus'].unique()
```

### LoanOriginationQuarter has the number of quarter and year so we split it into 2 columns

```python
cat.head(2)
```

```python
cat['Quarter_Number'] = cat['LoanOriginationQuarter'].str.split(' ').str.get(0)
cat['LoanOriginationQuarter'] = cat['LoanOriginationQuarter'].str.split(' ').str.get(1)
```

```python
cat.head(2)
```

### EmploymentStatus has a lot of values so we restrict it int 3 values Empolyed ,Not Employed,Retired

```python
cat['EmploymentStatus'].unique()
```

```python
cat['EmploymentStatus'] = cat['EmploymentStatus'].replace({
    'Self-employed':'Employed',
    'Full-time':'Employed',
    'Other':'Employed',
    'Part-time':'Employed',
    'Not available':'Not employed',
    np.nan: 'Not employed'
})
```

### one-hot encoding
### Apply one-hot encoding for multi-category columns to convert them into numeric features

```python
#  One Hot Encoding is a method for converting categorical variables into a binary format. 
#  It creates new columns for each category where 1 means the category is present and 0 means it is not. 
#  The primary purpose of One Hot Encoding is to 
#  ensure that categorical data can be effectively used in machine learning models.
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# creating one hot encoder object
onehotencoder = OneHotEncoder()

# reshape the 1-D country array to 2-D as fit_transform expects 2-D and fit the encoder
X = onehotencoder.fit_transform(cat.Quarter_Number.values.reshape(-1, 1)).toarray()
```

```python
# Convert the Transformed Data into a DataFrame and Concatenate it with the Original DataFrame

# Creating a DataFrame with the encoded data
dfOneHot = pd.DataFrame(X, columns=["Quarter_Number_" + str(int(i)) for i in range(X.shape[1])])

# Concatenating the original DataFrame with the encoded DataFrame
cat = pd.concat([cat, dfOneHot], axis=1)
```
